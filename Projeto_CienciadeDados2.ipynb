1. O tipo de pista e a condição metereológica interferem no acontecimento dos acidentes?
2. Qual estado apresenta maior quantidade de mortes e feridos?
3. Qual dia da semana ocorre a maior quantidade de acidentes?
4. Quais as causas mais prováveis para que ocorra acidentes?
5. Quais marcas de veículos estão envolvidas em mais acidentes?
6. Como é a distribuição de acidentes por idade dos envolvidos?
7. Quais são os mais envolvidos em acidentes? Homens ou mulheres?
8. Como o tipo de veículo pode interferir nas causas de acidentes?

> É importante frisar que as tabelas apresentam dados iguais a '(null)' como também dados nulos (NaN), dessa forma, até aqui nós transformamos os '(null)' dos atributos numéricos para NaN e apagamos tais linhas. Agora iremos buscar os dados não numéricos e transformar os '(null)' para uma mensagem de dado 'Não definido'.

```
#Altera o valor '(null)' por um mensagem de não informado
def altera_null(data, ano):
  if ano < 2016:
    result1 = data[['id','data_inversa','dia_semana','horario','br','km']]

    result2 = data[['uf','municipio','causa_acidente','tipo_acidente','classificacao_acidente','fase_dia','sentido_via',
                    'condicao_metereologica','tipo_pista','tracado_via','uso_solo']].applymap(lambda item: item.replace('(null)','Valor não definido'))
    
    result3 = data[['ano','pessoas','mortos','feridos_leves','feridos_graves',
                    'ilesos','ignorados','feridos','veiculos']]

    resultado1 = pd.merge(result1, result2, left_index=True, right_index=True)
    resultado2 = pd.merge(resultado1, result3, left_index=True, right_index=True)
    return resultado2

  else:
    result1 = data[['id','data_inversa','dia_semana','horario', 'br']]
    result2 = data[['km']].applymap(lambda item: item.replace(",","."))

    result3 = data[['uf','municipio','causa_acidente','tipo_acidente','classificacao_acidente','fase_dia','sentido_via',
                    'condicao_metereologica','tipo_pista','tracado_via','uso_solo']].applymap(lambda item: item.replace('(null)','Valor não definido'))
    
    result4 = data[['pessoas','mortos','feridos_leves','feridos_graves','ilesos','ignorados','feridos',
                    'veiculos']]

    resultado1 = pd.merge(result1, result2, left_index=True, right_index=True)
    resultado2 = pd.merge(resultado1, result3, left_index=True, right_index=True)
    resultado3 = pd.merge(resultado2, result4, left_index=True, right_index=True)
    return resultado3


ocorrencias_2007_new = altera_null(ocorrencias_2007_NOT_NaN, 2007)
ocorrencias_2008_new = altera_null(ocorrencias_2008_NOT_NaN, 2008)
ocorrencias_2009_new = altera_null(ocorrencias_2009, 2009)
ocorrencias_2010_new = altera_null(ocorrencias_2010, 2010)
ocorrencias_2011_new = altera_null(ocorrencias_2011, 2011)
ocorrencias_2012_new = altera_null(ocorrencias_2012_NOT_NaN, 2012)
ocorrencias_2013_new = altera_null(ocorrencias_2013, 2013)
ocorrencias_2014_new = altera_null(ocorrencias_2014, 2014)
ocorrencias_2015_new = altera_null(ocorrencias_2015, 2015)
ocorrencias_2016_new = altera_null(ocorrencias_2016_NOT_NaN, 2016)
ocorrencias_2017_new = altera_null(ocorrencias_2017_NOT_NaN, 2017)
ocorrencias_2018_new = altera_null(ocorrencias_2018_NOT_NaN, 2018)
ocorrencias_2019_new = altera_null(ocorrencias_2019_NOT_NaN, 2019)
ocorrencias_2020_new = altera_null(ocorrencias_2020_NOT_NaN, 2020)
```

```
data=df.sample(n=150000,random_state=42)
```

```
data['pessoas'].describe()
```

Aqui definimos a nossa metrica de avaliação da regressão. Optamos pela Raiz do erro quadratico medio (RMSE), que segue a mesma ideia do Erro Quadratico Medio (MSE), de penalisar diferenças muito significativas entre a amostra e o valor previsto. Isso é importante porque precimos ter uma precisão muito alta já que se trata de casos que envolvem risco a vida.

  Também aplicamos k-fold para dividir os dados em treino e teste para diminuirmos a influencia da ordem dos dados e termos resultados que representem de maneira mais segura e confiável os resultados dos algoritmos.

  Ao final, calculamos a media do RMSE que obtivemos de cada parte e com isso obtivemos o desempenho do algoritmo. 
      

```
from sklearn.metrics import mean_squared_error
```

```
#separando os dados de treino e teste
from sklearn.model_selection import KFold

k_fold=KFold(10)
indices_treino=[]
indices_teste=[]
for x,y in k_fold.split(var_independentes,var_dependente):
    indices_treino.append(x)
    indices_teste.append(y)
```

Os Algoritmos testados e os seus respectivos resultados foram:

* Regressão Linear: 1.41
* Arvore de Decisão: 1.40
* Floresta Aleatoria: 1.40
* MLP: 1.40


### Regressão Linear 

```
from sklearn.linear_model import LinearRegression
```

```
Lin_Reg=LinearRegression()

erro=[]

for ind_treino,ind_teste in zip(indices_treino,indices_teste):
    #Seperando os dados de treino e teste
    X_train,X_test=var_independentes.iloc[ind_treino],var_independentes.iloc[ind_teste]
    y_train,y_test=var_dependente.iloc[ind_treino],var_dependente.iloc[ind_teste]

    #treinando o modelo 
    Lin_Reg.fit(X_train,y_train)

    #As predicoes do modelo
    y_pred_Lin=Lin_Reg.predict(X_test)
    e=mean_squared_error(y_test,y_pred_Lin,squared=False)
    
    erro.append(e)

print('desempenho: {:.2f}'.format(np.mean(erro)))
```

Para avaliar o desempenho do nosso algoritmo, nos usamos o K-fold e a acuracia. O K-fold estabelecemos o K=3, já que era a quantidade de classes que tinhamos. Para medir o desempenho em cada parte de treino/teste usamos a acuracia.

Com isso, o desempenho do modelo ficou determinado como sendo a media da acuracia de cada parte.

```
#separando os dados de treino e teste
from sklearn.model_selection import KFold

k_fold=KFold(3)
indices_treino=[]
indices_teste=[]
for x,y in k_fold.split(caracteristica,classes):
    indices_treino.append(x)
    indices_teste.append(y)
```

### Modelos aplicados

Os modelos aplicados com os seus respectivos desempenho seguem abaixo:
* Arvore de decisão : 78,87%;
* MLP : 79,02%;
* K-nn: 76,31%;
* Regressão Logistica:76,13%; 

#### Arvore de decisão

```
from sklearn.tree import DecisionTreeClassifier
```

```
Tree_clf=DecisionTreeClassifier(max_depth=13)

acuracia_med=[]

for ind_treino,ind_teste in zip(indices_treino,indices_teste):
    #Seperando os dados de treino e teste
    X_train,X_test=caracteristica.iloc[ind_treino],caracteristica.iloc[ind_teste]
    y_train,y_test=classes.iloc[ind_treino],classes.iloc[ind_teste]

    #treinando o modelo 
    Tree_clf.fit(X_train,y_train)

    #As predicoes do modelo
    acuracia=Tree_clf.score(X_test,y_test)
    
    
    acuracia_med.append(acuracia)

print('Desempenho:{:.2f}%'.format(np.mean(acuracia_med)*100))

```

```
from sklearn.neural_network import MLPClassifier
```

```
MLP_clf=MLPClassifier((300,100,50),activation='relu',learning_rate_init=0.0001,verbose=True,early_stopping=True,n_iter_no_change=20)

acuracia_med=[]

for ind_treino,ind_teste in zip(indices_treino,indices_teste):
    #Seperando os dados de treino e teste
    X_train,X_test=caracteristica.iloc[ind_treino],caracteristica.iloc[ind_teste]
    y_train,y_test=classes.iloc[ind_treino],classes.iloc[ind_teste]

    #treinando o modelo 
    MLP_clf.fit(X_train,y_train)

    #As predicoes do modelo
    acuracia=MLP_clf.score(X_test,y_test)
    
    
    acuracia_med.append(acuracia)

print('Desempenho:{:.2f}%'.format(np.mean(acuracia_med)*100))
```
